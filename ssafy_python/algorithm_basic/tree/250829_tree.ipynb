```python
icp = {'(': 3, ')': 3, '*': 2, '/': 2, '+': 1, '-': 1} # 스택에 푸쉬할때 읽은 우선순위
isp = {'(': 0, ')': 3, '*': 2, '/': 2, '+': 1, '-': 1} # 스택 내부에서의 우선순위

infix = '(6+5*(2-8)/2+6)'   # 중위 표기
postfix = ''                # 후위 표기
S = []

# 중위 표기에서 token을 하나씩 읽어서 반복
for token in infix:
    # 연산자면
    if token in icp:
        # token이 ')' 면 S 에 '('가 나올 때까지 pop해서 출력하고 마지막으로 '('를 pop
        if token == ')':
            while S and S[-1] != '(': # S: 스택을 사용할 때는 항상 빈 스택인지 확인!
                postfix += S.pop()
            S.pop() # 마지막으로 '('를 pop
        else:
            # 나머지는 S에 push, token보다 우선순위(먼저 계산되는 순)가 높은 연자자들을 모두 S에서 pop해서 출력
            while S and isp[S[-1]] >= icp[token]:
                postfix += S.pop()
            # token push
            S.append(token)
    else: # 피연산자면 postfix에 출력
        postfix += token

while S:
    postfix += S.pop()

print(postfix)
```
