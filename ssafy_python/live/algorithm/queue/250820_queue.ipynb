```python
# 원형 큐의 구현
n = 5
# 1. 초기 공백 큐 생성
    # front와 rear을 0 으로 초기화
cq = [0] * n
front = rear = 0

# 2. 삽입 enqueue(item)
    # rear값을 조정해 새로운 원소를 삽입할 자리 지정
def enqueue(item):
    global rear
    if is_full(): print('Queue_Full')
    else: 
        rear = (rear + 1) % n
        cq[rear] = item
    
# 3. 삭제 dequeue()
    # front값을 조정해 삭제할 자리 지정
def dequeue():
    global front
    if is_empty(): return 'Queue_Empty'
    else:
        front = (front + 1) % n
        return cq[front]

# 4. 공백상태 및 포화상태 검사 is_empty(), is_full()
    # 공백상태: front == rear
    # 포화상태: (rear + 1) % n == front
def is_empty():
    return front == rear

def is_full():
    return (rear + 1) % n == front

# 5. 코드 실행
for i in range(1, 7):
    if i % 2 == 1:
        print('dequeue: ', dequeue())
        print('front:', front, 'rear:', rear)
        print(cq)
    enqueue(i)
    print(i, cq)
```

```python
# 연결 큐의 구현 => 이해 아직 못함

class Node:
    def __init__(self, item, n=None):
        self.item = item # 노드가 저장하는 값
        self.next = n # 다음 노드를 가리키는 포인터(링크)

def enqueue(item): # 연결 큐의 삽입 연산
    global front, rear
    newNode = Node(item) # 새로운 노드 생성
    if front == None: # 큐가 비어있다면
        front = newNode
    else: # 마지막 노드(rear)가 새 노드를 가리키도록 연결하는 코드
        rear.next = newNode
    rear = newNode

def is_empty():
    return front == None

def dequeue(): # 연결 큐의 삭제 연산
    global front, rear
    if is_empty():
        print('Queue_Empty')
        return None
    
    item = front.item
    front = front.next
    if front == None:
        rear = None
    return item

def qpeek():
    return front.item

def print_q():
    f = front
    s = ''
    while f:
        s += f.item + ''
        f = f.next
    return s

front = None
rear = None
```
